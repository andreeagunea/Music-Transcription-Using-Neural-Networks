```python
# K-means pe CSV notetrainsitest cu stergere note gresite, comparatie diferiti algoritmi Kmeans

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
def identificate_gresit(nume_fisiere, y):
    linii_gresite=[]
    linii = nume_fisiere.shape[0]                           
    for i in range(linii):                                 
        if nume_fisiere[i][2] == "#":
            nota = nume_fisiere[i][:3]                      
        else:
            nota = nume_fisiere[i][:2]                     
        if sorted(y[i]) != sorted(nota):                    
            linii_gresite.append(i+2)                       
            print("Nota identificata gresit in pe randul {}, nota adevarata {}, nota falsa {}.".format(i+2, nota, y[i]))
    print("\nNumar note identificate gresit: {}.".format(len(linii_gresite)))
    
    return linii_gresite


def stergere_clase(nume_fisiere, X, y, linii_gresite):
    indecsi_gresiti = [i - 2 for i in linii_gresite]
    nume_fisiere = np.delete(nume_fisiere, indecsi_gresiti)
    X = np.delete(X, indecsi_gresiti)
    y = np.delete(y, indecsi_gresiti)
    print("\nClase gresite sterse.")
             
    return nume_fisiere, X, y
```

```python
df = pd.read_csv('autocor_trainsitest.csv')
nume_fisiere = df.iloc[:, 0].values
frecvente = df.iloc[:, 1].values
note = df.iloc[:, 2].values

X = frecvente
X = X.reshape(-1,1)
kmeans = KMeans(n_clusters=37)
kmeans.fit(X)
identified_cluters = kmeans.fit_predict(X)

data_with_clusters = df.copy()
data_with_clusters['Clusters'] = identified_cluters
indici = np.arange(1, len(X) + 1)
plt.figure(1)
plt.scatter(indici, data_with_clusters['pitch'],
                c=data_with_clusters['Clusters'], cmap='rainbow')
plt.show()
```

```python
df = pd.read_csv('autocor_trainsitest.csv')
nume_fisiere = df.iloc[:, 0].values
frecvente = df.iloc[:, 1].values
note = df.iloc[:, 2].values

X = frecvente
y = note

linii_gresite = identificate_gresit(nume_fisiere, y)
nume_fisiere, X, y = stergere_clase(nume_fisiere, X, y, linii_gresite)
linii_gresite = identificate_gresit(nume_fisiere, y)

# refacere dataset de note corecte
column_values = ['fisier', 'pitch', 'nota']
index_values = [i for i in range(int(X.shape[0]))]
concat = np.column_stack((nume_fisiere, X, y))
df = pd.DataFrame(data=concat,
                  index=index_values,
                  columns=column_values)

X = X.reshape(-1,1)
kmeans = KMeans(n_clusters=37)
kmeans.fit(X)
identified_cluters = kmeans.fit_predict(X)

data_with_clusters = df.copy()
data_with_clusters['Clusters'] = identified_cluters
indici = np.arange(1, len(X) + 1)
plt.figure(2)
plt.scatter(indici, data_with_clusters['pitch'],
                c=data_with_clusters['Clusters'], cmap='rainbow')
plt.show()
```
