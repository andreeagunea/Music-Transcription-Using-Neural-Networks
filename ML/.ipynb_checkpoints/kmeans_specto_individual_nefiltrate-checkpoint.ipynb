```python
# K-means pe SPECTOGRAME FILTRATE notetrainsitest, comparatie diferiti algoritmi Kmeans

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans

from k_means_constrained import KMeansConstrained

import os
import cv2
```

```python
def erori_kmeans(note, clustere):
    nota_curenta = None
    cluster_curent = None
    nr = 0
    for i in range(note.shape[0]):
        if nota_curenta != note[i]:
            nota_curenta = note[i]
            cluster_curent = clustere[i]
        if cluster_curent != clustere[i]:
            nr += 1
            print("Eroare K-means in CSV pe randul {}".format(i+2))
    print("{} erori".format(nr))
    
    return nr
```

```python
# Date antrenare
folder = 'spectograme_trainsitest'
path_folder = os.path.join(os.getcwd(), folder)
fisiere = os.listdir(path_folder)
path_fisiere = [os.path.join(path_folder, i) for i in fisiere]

imagini = [cv2.imread(i) for i in path_fisiere]
imagini = np.array(imagini)
dim = len(imagini[0]) * len(imagini[0][0]) * 3
imagini = imagini.reshape((imagini.shape[0]), dim)
```

```python
X = imagini
kmeans = KMeans(n_clusters=37)
kmeans.fit(X)
clustere = kmeans.fit_predict(X)
```

```python
note = [i[:3] if i[2]=='#' else i[:2] for i in fisiere]
note = np.array(note)
```

```python
nr = erori_kmeans(note, clustere)
acuratete = (imagini.shape[0] - nr)/imagini.shape[0]*100
print("\nAcuratete: {}%".format(acuratete))
```
