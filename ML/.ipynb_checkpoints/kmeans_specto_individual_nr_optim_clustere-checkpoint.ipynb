```python
# K-means pe SPECTOGRAME FILTRATE notetrainsitest, comparatie diferiti algoritmi Kmeans

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans

import os
import cv2
from sklearn.preprocessing import LabelEncoder

from sklearn.metrics import accuracy_score

from sklearn.cluster import MiniBatchKMeans


def extragere_date(folder):
    path_folder = os.path.join(os.getcwd(), folder)
    fisiere = os.listdir(path_folder)
    path_fisiere = [os.path.join(path_folder, i) for i in fisiere]

    imagini = [cv2.imread(i) for i in path_fisiere]
    imagini = np.array(imagini)
    dim = len(imagini[0]) * len(imagini[0][0]) * 3
    imagini = imagini.reshape((imagini.shape[0]), dim)

    X = imagini
    note = [i[:3] if i[2]=='#' else i[:2] for i in fisiere]
    note = np.array(note)
    le = LabelEncoder()
    clase = le.fit_transform(note)
    

    return X, clase
    
def kmeans_acuratete(clustere, clase):
    dictionar = {}
    clustere_unice = np.unique(clustere)
    for i in range(len(clustere_unice)):
        masca = np.where(clustere == i, 1, 0)
        aparitii_etichete = np.bincount(clase[masca==1])
        eticheta_aleasa = aparitii_etichete.argmax()
        dictionar[i] = eticheta_aleasa

    clustere_clase = np.random.rand(len(clustere))
    for i in range(len(clustere)):
        clustere_clase[i] = dictionar[clustere[i]]

    acuratete = accuracy_score(clustere_clase, clase)
    
    return acuratete, clustere_clase


# def plot_clusters():
#     # Plot cu frecvente(indici).
#     indici = np.arange(1, len(X) + 1)
#     plt.figure(1)
#     plt.scatter(indici, X)

#     X = X.reshape(-1,1)
#     kmeans = KMeans(n_clusters=37)
#     kmeans.fit(X)

#     identified_cluters = kmeans.fit_predict(X)
#     # print(identified_cluters)

#     data_with_clusters = dataset.copy()
#     data_with_clusters['Clusters'] = identified_cluters
#     # print(data_with_clusters)
#     print(data_with_clusters['Clusters'])

#     # Plot cu clusterele pentru frecvente(indici).
#     plt.figure(2)
#     plt.scatter(indici, data_with_clusters['note_freq'],
#                 c=data_with_clusters['Clusters'], cmap='rainbow')
#     plt.show()
```

```python
nr_clustere = [37, 40, 45, 49, 55, 60]
iteratii = 10
scoruri_testbench = []

for i in range(1, iteratii+1):
    print("\nIteratia numarul ", str(i))
    scoruri_iteratie = []
    for i in nr_clustere:
        kmeans = MiniBatchKMeans(n_clusters=i)
        kmeans.fit(X)
        clustere = kmeans.fit_predict(X)

        reference_labels = {}
        for j in range(len(np.unique(kmeans.labels_))):
            index = np.where(kmeans.labels_ == j,1,0)
            num = np.bincount(clase[index==1]).argmax()
            reference_labels[j] = num

        number_labels = np.random.rand(len(kmeans.labels_))
        for j in range(len(kmeans.labels_)):
          number_labels[j] = reference_labels[kmeans.labels_[j]]

        acuratete, clustere_clase = kmeans_acuratete(clustere, clase)
        print('Scor', i, 'clustere:', acuratete)
        scoruri_iteratie.append(acuratete)
        
    scoruri_testbench.append(scoruri_iteratie)
    
    
d = dict()
for i in range(len(nr_clustere)):
    cheie = str(nr_clustere[i]) 
    d[cheie] = []
    
    for j in range(len(scoruri_testbench)):
        d[cheie].append(scoruri_testbench[j][i])
    
    suma_scoruri = sum(d[cheie])
    scor_mediu = suma_scoruri / len(d[cheie])

    print('\nScor mediu', cheie, 'clustere:', scor_mediu)
```
